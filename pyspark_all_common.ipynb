# Custom Schema Data Frame

```python
# prompt: write a code to create a custome schema for a dataset n pyspark

from pyspark.sql.types import StructType, StructField, IntegerType, StringType

data = [
    (1, "Alice", 25),
    (2, "Bob", 30),
    (3, "Charlie", 28)
]

# Define the schema
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])

# Create a DataFrame with the custom schema
df_custom_schema = spark.createDataFrame(data, schema)

# Show the DataFrame
df_custom_schema.show()
```

# Missing Data Handling

```python
# prompt: write down code for all the methods to deal with the missing data in pyspark

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, mean

# Sample data with missing values
data = [
    (1, "Alice", 25),
    (2, "Bob", None),
    (3, None, 28),
    (4, "David", 32),
    (5, "Emily", None)
]

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])

spark = SparkSession.builder.appName("MissingDataHandling").getOrCreate()
df = spark.createDataFrame(data, schema)

# 1. Drop rows with missing values
df_dropped = df.na.drop()
df_dropped.show()

# 2. Fill missing values with a constant
df_filled_constant = df.na.fill({"age": 0, "name": "Unknown"})
df_filled_constant.show()

# 3. Fill missing values with mean
mean_age = df.select(mean(col("age"))).collect()[0][0]
df_filled_mean = df.na.fill({"age": mean_age})
df_filled_mean.show()

# 4. Replace missing values using a conditional expression
df_replaced = df.withColumn("age", when(col("age").isNull(), 0).otherwise(col("age")))
df_replaced.show()

spark.stop()
```
