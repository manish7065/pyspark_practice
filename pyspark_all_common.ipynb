# Missing Data Handling

```python
# prompt: write down code for all the methods to deal with the missing data in pyspark

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, mean

# Sample data with missing values
data = [
    (1, "Alice", 25),
    (2, "Bob", None),
    (3, None, 28),
    (4, "David", 32),
    (5, "Emily", None)
]

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])

spark = SparkSession.builder.appName("MissingDataHandling").getOrCreate()
df = spark.createDataFrame(data, schema)

# 1. Drop rows with missing values
df_dropped = df.na.drop()
df_dropped.show()

# 2. Fill missing values with a constant
df_filled_constant = df.na.fill({"age": 0, "name": "Unknown"})
df_filled_constant.show()

# 3. Fill missing values with mean
mean_age = df.select(mean(col("age"))).collect()[0][0]
df_filled_mean = df.na.fill({"age": mean_age})
df_filled_mean.show()

# 4. Replace missing values using a conditional expression
df_replaced = df.withColumn("age", when(col("age").isNull(), 0).otherwise(col("age")))
df_replaced.show()

spark.stop()
```

# Optimization in pyspark

## Broadcast Joint

Problem Statement: Joining a large dataset with a smaller dataset can be slow due to the shuffle operation.

Solution: Use broadcast join to broadcast the smaller dataset to all nodes, avoiding the shuffle.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import broadcast

spark = SparkSession.builder.appName("BroadcastJoin").getOrCreate()

large_df = spark.range(0, 100000000).toDF("id")

samll_df = spark.createDataFrame([(1,"manish"),(2,'subham'),(3,'prateek')],["id",'name'])

result_df = large_df.join(broadcast(samll_df), large_df.id == samll_df.id)

result_df.show()

# spark.stop()
```

## caching

Problem Statement: Repeated operations on the same DataFrame can lead to redundant computation.

Solution: Cache the DataFrame to reuse it without recomputing.

```python
df = spark.range(0, 100000000).toDF("id")
transformed_df = df.withColumn("id2", df['id']+1)

transformed_df.cache()
transformed_df.count()

transformed_df.show()
```

## Partitioning

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import time

# Initialize Spark session
spark = SparkSession.builder.appName("PartitioningExample").getOrCreate()

# Create a sample DataFrame with sales data
data = [(1, '2023-01-01', 'A', 100),
        (2, '2023-01-02', 'B', 200),
        (3, '2023-01-03', 'A', 300),
        (4, '2023-01-04', 'B', 400),
        (5, '2023-01-05', 'A', 500),
        (6, '2023-01-06', 'B', 600),
        (7, '2023-01-07', 'A', 700),
        (8, '2023-01-08', 'B', 800)]

columns = ["id", "date", "category", "amount"]

df = spark.createDataFrame(data, columns)
df.show()
```

### without partitioning

```python
# Measure time for aggregation without partitioning
start_time = time.time()
result_df = df.groupBy("category").sum("amount")
result_df.show()
print(f"Aggregation without partitioning: {time.time() - start_time} seconds")
```

#### with partitioning

```python
# Repartition the DataFrame by category
partitioned_df = df.repartition("category")

# Measure time for aggregation with partitioning
start_time = time.time()
partitioned_result_df = partitioned_df.groupBy("category").sum("amount")
partitioned_result_df.show()
print(f"Aggregation with partitioning: {time.time() - start_time} seconds")
```

### Other Example

```python
from pyspark.sql import SparkSession
import time

# Initialize Spark session
spark = SparkSession.builder.appName("PartitioningJoinExample").getOrCreate()

# Sample customer data
customer_data = [(1, 'Alice'),
                 (2, 'Bob'),
                 (3, 'Charlie'),
                 (4, 'David'),
                 (5, 'Eve'),
                 (6, 'Frank'),
                 (7, 'Grace'),
                 (8, 'Heidi')]

customer_columns = ["customer_id", "name"]

customers_df = spark.createDataFrame(customer_data, customer_columns)

# Sample order data
order_data = [(1, '2023-01-01', 100),
              (2, '2023-01-02', 150),
              (1, '2023-01-03', 200),
              (3, '2023-01-04', 250),
              (2, '2023-01-05', 300),
              (4, '2023-01-06', 350),
              (5, '2023-01-07', 400),
              (3, '2023-01-08', 450),
              (6, '2023-01-09', 500),
              (7, '2023-01-10', 550),
              (8, '2023-01-11', 600)]

order_columns = ["customer_id", "order_date", "amount"]

orders_df = spark.createDataFrame(order_data, order_columns)
```

```python
# Measure time for join without partitioning
start_time = time.time()
joined_df = customers_df.join(orders_df, "customer_id")
joined_df.show()
print(f"Join without partitioning: {time.time() - start_time} seconds")
```

```python
# Repartition both DataFrames by customer_id
partitioned_customers_df = customers_df.repartition("customer_id")
partitioned_orders_df = orders_df.repartition("customer_id")

# Measure time for join with partitioning
start_time = time.time()
partitioned_joined_df = partitioned_customers_df.join(partitioned_orders_df, "customer_id")
partitioned_joined_df.show()
print(f"Join with partitioning: {time.time() - start_time} seconds")
```
